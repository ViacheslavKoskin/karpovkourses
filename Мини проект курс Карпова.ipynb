```python
import pandas as pd
```

```python
bookings = pd.read_csv('2_bookings.csv', encoding = 'UTF-8', sep = ';')
```

```python
import pandas as pd

s = pd.Series(bookings.columns)
v = s.str.lower()

b = v.str.replace(' ', '_')

names = {}
num = 0
for i in s:
    names[i] = b[num]
    num = num + 1

bookings = bookings.rename(columns=names)

bookings
```

```python
bookings.columns
```

```python
#рейтинг стран по количеству броней

country_rate = bookings \
    .query("is_canceled == 0") \
    .country.value_counts()

country_rate
```

```python
#посчитай среднее значение бронирований на ночь по типам отелей

type_mean = bookings \
    .groupby('hotel') \
    .aggregate({'stays_total_nights':'mean'})

type_mean.round(2)
```

```python
#найди те брони, где название номера отличается от изначально забронированного

rooms = bookings \
    .query("assigned_room_type != reserved_room_type") 

rooms.shape
```

```python
#какой месяц был самым популярным?

arrival_date_month_rate = bookings \
    .query("arrival_date_year == 2017") \
    .arrival_date_month.value_counts()

arrival_date_month_rate
```

```python
bookings['has_kids'] = 0
```

```python
bookings['has_kids'] = bookings.total_kids != 0
bookings['has_kids']    
```

```python
Group_True = bookings.query("has_kids == True") \
                    .agg({'has_kids':'count'})
```

```python
ChUrn_group_True = bookings.query("has_kids == True") \
                           .query("is_canceled != 0") \
                           .agg({'has_kids':'count'})
```

```python
churn_True =  ChUrn_group_True/Group_True
churn_True
```

```python
Group_False = bookings.query("has_kids == False") \
                     .agg({'has_kids':'count'}) 

ChUrn_group_False = bookings.query("has_kids == False") \
                           .query("is_canceled != 0") \
                           .agg({'has_kids':'count'})

churn_False =  ChUrn_group_False/Group_False

churn_False
```
